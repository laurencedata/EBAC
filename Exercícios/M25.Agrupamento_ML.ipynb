## 3\. Modelagem 

### **3.1. Treino** 

Treine 10 modelos de **k-médias** variando o número de clusters de 1 a 10. Para cada modelo treinado, salve o valor global do **wcss** em uma lista.

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans()
model = KMeans(n_clusters=5)
model = model.fit(iris)
```

```python
model.__dict__
```

```python
clusters = model.labels_
clusters = pd.DataFrame(clusters, columns=["cluster"])
clusters.head()
```

```python
clustered_iris = pd.concat([iris, clusters], axis=1)
clustered_iris.head()
```

```python
iris.shape
```

```python
with sns.axes_style("whitegrid"):
    grafico = sns.pairplot(data=clustered_iris, hue="cluster", palette="pastel")
```

Verificando a posição de cada centróide:

```python
clusters_centers = model.cluster_centers_
clusters_centers = pd.DataFrame(clusters_centers, columns=["sepal_length", "sepal_width", "petal_length", "petal_width"])
clusters_centers.head()
```

```python
with sns.axes_style("whitegrid"):
    fig, ax = plt.subplots()
    sns.scatterplot(data=clustered_iris, x="sepal_length", y="sepal_width", hue="cluster", palette="pastel", ax=ax)
    sns.scatterplot(data=clusters_centers, x="sepal_length", y="sepal_width", color="black", ax=ax)
```

```python
with sns.axes_style("whitegrid"):
    fig, ax = plt.subplots()
    sns.scatterplot(data=clustered_iris, x="petal_length", y="petal_width", hue="cluster", palette="pastel", ax=ax)
    sns.scatterplot(data=clusters_centers, x="petal_length", y="petal_width", color="black", ax=ax)
```

### **3.2. Avaliação** 

Gere um gráfico de linha dos valores do **wcss** pelo **número de clusters**. Utilize o método do cotovelo para decidir o número final de clusters.

```python
wcss = []

for k in range(1, 11):
    model = KMeans(n_clusters=k)
    model = model.fit(iris)
    wcss.append(model.inertia_)
```

```python
with sns.axes_style("whitegrid"):
    grafico = sns.lineplot(x=range(1,11), y=wcss, marker="8", palette="pastel")
    grafico.set(title="Método do Cotovelo", ylabel="WCSS", xlabel="Qtd. Clusters")
```

### **3.3. Visualização** 

**a)** Utilizando o número de clusters final, adicione uma coluna chamada **cluster** no dataframe **iris** com o número do cluster que cada flor foi alocada.

```python
model = KMeans(n_clusters=3)
model = model.fit(iris)

clusters = model.labels_
clustered_iris = pd.concat([iris, pd.DataFrame(clusters, columns=["clusters"])], axis=1)

with sns.axes_style("whitegrid"):
    grafico = sns.pairplot(data=clustered_iris, hue="clusters", palette="pastel")
```
