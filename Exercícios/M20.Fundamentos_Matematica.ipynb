### **1.1. Nativo** 

Neste primeira atividade, você deve ler o código abaixo implementado em Python nativo e responder a seguinte pergunta:

> O que o código abaixo computa?

**Resposta**: 

Analisando o código abaixo, a estrutura utilizada com todos os recursos são pela própria linguagem Python, sem nenhuma biblioteca ou pacote externo.

O código começa abrindo o arquivo 'traffic.csv', no modo leitura. Após a leitura e o carregamento dos dados, parte para a análise, fazendo a filtragem começando no dia 14, com o número de incidentes e incidentes por dia.

Depois faz um looping, nomeando de 'timebox', que significa, caixa de tempo. Ou seja, quer dizer que quer retornar todas as horas desse dia dentro da variável data, utilizando a função split para carregar nova linha e cada hora, separado por ponto e vírgula (';').

E assim segue nas demais variáveis, incidents e o resultado final, que é 'incidente por dia'.

Depois disso, utiliza-se a função try para contabilizar os restantes dos dias, além do 14. São 5 dias, os intervalos de horas vão até o número 27, a partir daí, contabiliza do dia 14 até o dia 18.

No final, e o principal objetivo do código abaixo é computar o total de acidentes por dia.

14, 15, 16, 17 e 18 são os dias que constam no arquivo 'traffic.csv'. 

16, 63, 94, 82 e 85 é o total de acidentes que teve durante a cada dia.



```python
# -- read

data = None

with open(file='traffic.csv', mode='r', encoding='utf8') as fp:
  
  fp.readline()
  data = fp.read()

# -- analytics

day = 14
incidents = 0
incident_by_day = dict()

for timebox in data.split(sep='\n'):

  timebox_data = timebox.split(sep=';')

  # --
  # -- inicio da computação escalar
  # --

  for incident in timebox_data[1: len(timebox_data)-1]:
    incidents = incidents + int(incident)

  # --
  # -- fim da computação escalar
  # --

  try: 

    half_hour = int(timebox_data[0])

    if half_hour == 27:
      incident_by_day[day] = incidents
      day = day + 1
      incidents = 0

  except ValueError:
    continue

# -- results

for day in incident_by_day:
  
  print(f'{day}: {incident_by_day[day]}')
```

### **1.2. NumPy** 

Substitua o trecho do código do algoritmo que utiliza da **computação escalar** por um que utiliza da **computação vetorial**. Use o pacote NumPy.

---

```python
import numpy as np

# -- read

data = None

with open(file='traffic.csv', mode='r', encoding='utf8') as fp:
    fp.readline()
    data = fp.read()

# -- analytics

day = 14
incidents = np.zeros(16, dtype=int)
incident_by_day = dict()

for timebox in data.split(sep='\n'):
    if not timebox:
        continue  # ignora linhas vazias

    timebox_data = timebox.split(sep=';')

    # verifica se a linha tem o número esperado de elementos
    if len(timebox_data) != 18:
        continue  # ignora linhas com formato incorreto

    # --
    # -- inicio da computação vetorial
    # --

    incidents += np.array(timebox_data[1: len(timebox_data)-1], dtype=int)

    # --
    # -- fim da computação vetorial
    # --

    try:
        half_hour = int(timebox_data[0])

        if half_hour == 27:
            incident_by_day[day] = np.sum(incidents)
            day += 1
            incidents = np.zeros(16, dtype=int)  # reinicializa para o próximo dia

    except ValueError:
        print(f"Erro de valor em '{timebox_data[0]}'")
        continue

# -- results

for day in incident_by_day:
    print(f'{day}: {incident_by_day[day]}')
```
