**sqv** = Quantidade Total das Vendas;

**tkt** = Ticket Médio.

## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```python
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'
```

```python
# Método 1 - Função que contabiliza a quantidade de caracteres dentro de uma string
print(cancao.find("roda gigante"))
# Método 2 - Função que possibilita substituir palavras, siglas, etc
print(cancao.replace("roda gigante", "RG"))
# Método 3 - Função que transforma a string em caixa alta
print(cancao.upper())
```

Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```python
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'
```

```python
taxa_selic = noticia.find("2,75%")
# Descobrindo a posição inicial
print('Taxa Selic:', taxa_selic)
# Aplicando o fatiamento fixo
print('Selic:', noticia[12:18])
```
